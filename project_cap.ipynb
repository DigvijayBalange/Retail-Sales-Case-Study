```python
import json
import pandas as pd

# Open the JSON file
with open('behaviour.json', 'r') as file:
    # Read the contents of the file
    data = file.read()


# Convert JSON string to Python object
json_data = json.loads(data)

# Initialize an empty DataFrame with the desired columns
df_beh = pd.DataFrame(columns=["ID","Recency","MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts",
                           "MntGoldProds","NumDealsPurchases","NumWebPurchases","NumCatalogPurchases","NumStorePurchases",
                           "NumWebVisitsMonth"])

# Iterate over each dictionary in the list
for d in json_data:
    # Extract the customer ID
    customer_id = list(d.keys())[0]
    
    # Extract the values for the customer attributes
    values = list(d.values())[0]
    
    # Append a new row to the DataFrame with the customer ID and attribute values
    df_beh = df_beh.append({"ID": customer_id, **values}, ignore_index=True)

# Display DataFrame
df_beh
```

```python
df_beh
```

```python
new_columns = df_beh['ID'].str.split('_', expand=True)
df_beh[['ID_obj', 'ID_int']] = new_columns
df_beh.info()
```

```python
# Assume df is your DataFrame
# Convert 'column_name' from object type to integer type
df_beh['Recency'] = df_beh['Recency'].astype(int)
df_beh['MntWines'] = df_beh['MntWines'].astype(int)
df_beh['MntFruits'] = df_beh['MntFruits'].astype(int)
df_beh['MntMeatProducts'] = df_beh['MntMeatProducts'].astype(int)
df_beh['MntFishProducts'] = df_beh['MntFishProducts'].astype(int)
df_beh['MntSweetProducts'] = df_beh['MntSweetProducts'].astype(int)
df_beh['MntGoldProds'] = df_beh['MntGoldProds'].astype(int)
df_beh['NumDealsPurchases'] = df_beh['NumDealsPurchases'].astype(int)
df_beh['NumWebPurchases'] = df_beh['NumWebPurchases'].astype(int)
df_beh['NumCatalogPurchases'] = df_beh['NumCatalogPurchases'].astype(int)
df_beh['NumStorePurchases'] = df_beh['NumStorePurchases'].astype(int)
df_beh['NumWebVisitsMonth'] = df_beh['NumWebVisitsMonth'].astype(int)
df_beh['ID_int'] = df_beh['ID_int'].astype(int)
df_beh.info()
```

### Hypothesis 1
Customers who accept Campaign 1 are more likely to respond to marketing campaigns compared to those who accept Campaign 2. 

H0: There is no significant difference in the response rate between customers who accepted Campaign 1 and those who accepted Campaign 2.

Ha: Customers who accepted Campaign 1 are more likely to respond to marketing campaigns compared to those who accepted Campaign 2.

```python
from scipy.stats import chi2_contingency

# Subset data for customers who accepted Campaign 1 and Campaign 2
campaign_1 = df_final[df_final['AcceptedCmp1'] == 1]
campaign_2 = df_final[df_final['AcceptedCmp2'] == 1]

# Create a contingency table
contingency_table = pd.crosstab(df_final['Response'], df_final['AcceptedCmp1'])

# Add row and column names for better readability
contingency_table.index = ['Not Responded', 'Responded']
contingency_table.columns = ['Not Accepted Campaign 1', 'Accepted Campaign 1']
print("Contingency Table:")
print(contingency_table)
print('--')
# Perform chi-square test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Define significance level
alpha = 0.05

# Print results
print("Hypothesis 1: Customers who accept Campaign 1 are more likely to respond to marketing campaigns compared to those who accept Campaign 2.")
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the response rate between customers who accepted Campaign 1 and those who accepted Campaign 2.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the response rate between customers who accepted Campaign 1 and those who accepted Campaign 2.")
```

### Hypothesis 2
The number of store purchases is positively associated with the response rate.

H0: There is no significant correlation between the number of store purchases and the response rate.

Ha: The number of store purchases is positively associated with the response rate.

```python
from scipy.stats import pearsonr

# Calculate correlation coefficient between 'NumStorePurchases' and 'Response'
correlation, p_value = pearsonr(df_final['NumStorePurchases'], df_final['Response'])

# Define significance level
alpha = 0.05

# Print results
print("Hypothesis 2: The number of store purchases is positively associated with the response rate.")
print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant correlation between the number of store purchases and the response rate.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation between the number of store purchases and the response rate.")
```

### Hypothesis 3
Customers who spend more on wines are more likely to respond to marketing campaigns.

H0: There is no significant difference in the amount spent on wines between responders and non-responders.

Ha: Customers who spend more on wines are more likely to respond to marketing campaigns.

```python
from scipy.stats import ttest_ind
# Subset data for responders and non-responders
responders = df_final[df_final['Response'] == 1]
non_responders = df_final[df_final['Response'] == 0]

# Perform t-test for amount spent on wines between responders and non-responders
t_stat, p_value = ttest_ind(responders['MntWines'], non_responders['MntWines'])

# Define significance level
alpha = 0.05

# Print results
print("Hypothesis 3: Customers who spend more on wines are more likely to respond to marketing campaigns.")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the amount spent on wines between responders and non-responders.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the amount spent on wines between responders and non-responders.")
```

### Hypothesis 4
The number of web visits per month is positively associated with the response rate.

H0: There is no significant correlation between the number of web visits per month and the response rate.

Ha: The number of web visits per month is positively associated with the response rate.

```python
from scipy.stats import spearmanr

# Calculate Spearman correlation coefficient and p-value
correlation, p_value = spearmanr(df_final['NumWebVisitsMonth'], df_final['Response'])

# Define significance level
alpha = 0.05

# Print results
print("Hypothesis 4: The number of web visits per month is positively associated with the response rate.")
print(f"Spearman correlation coefficient: {correlation}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant correlation between the number of web visits per month and the response rate.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation between the number of web visits per month and the response rate.")
```

## Testing the hypothesis that recent customers complain less in general compared to older customers?

```python
# Convert 'Joining_Date' column to datetime
df_final['Dt_Customer'] = pd.to_datetime(df_final['Dt_Customer'])
```

```python
df_final['Dt_Customer'].max()
```

```python
df_final['Dt_Customer'].min()
```

```python
mean_date = df_final['Dt_Customer'].mean()
mean_date
```

```python
# Determine recent and older customers
recent_customers = df_final[df_final['Dt_Customer'] >= mean_date]
older_customers = df_final[df_final['Dt_Customer'] < mean_date]

# Calculate complaint rates
recent_complaint_rate = recent_customers['Complain'].mean()
older_complaint_rate = older_customers['Complain'].mean()

from scipy.stats import ttest_ind
# Perform t-test
t_stat, p_value = ttest_ind(recent_customers['Complain'], older_customers['Complain'])

# Define significance level
alpha = 0.05

# Print results
print(f"Recent Customer Complaint Rate: {recent_complaint_rate:.2f}")
print(f"Older Customer Complaint Rate: {older_complaint_rate:.2f}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: Recent customers complain less in general compared to older customers.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in complaint rates between recent and older customers.")
```

## Profile of people who respond vs. who dont

```python
# Separate responders and non-responders
responders = df_final[df_final['Response'] == 1]
non_responders = df_final[df_final['Response'] == 0]

mean_income_responders = responders['Income_val'].mean()
mean_income_non_responders = non_responders['Income_val'].mean()

education_distribution_responders = responders['Education'].value_counts(normalize=True)
education_distribution_non_responders = non_responders['Education'].value_counts(normalize=True)

Marital_Status_distribution_responders = responders['Marital_Status'].value_counts(normalize=True)
Marital_Status_distribution_non_responders = non_responders['Marital_Status'].value_counts(normalize=True)

Kidhome_distribution_responders = responders['Kidhome'].value_counts(normalize=True)
Kidhome_distribution_non_responders = non_responders['Kidhome'].value_counts(normalize=True)

Teenhome_distribution_responders = responders['Teenhome'].value_counts(normalize=True)
Teenhome_distribution_non_responders = non_responders['Teenhome'].value_counts(normalize=True)

NumWebPurchases_distribution_responders = responders['NumWebPurchases'].value_counts(normalize=True)
NumWebPurchases_distribution_non_responders = non_responders['NumWebPurchases'].value_counts(normalize=True)

# Create a profile DataFrame
profile = pd.DataFrame(columns=['Characteristic', 'Responders', 'Non-Responders'])
profile = profile.append({'Characteristic': 'Mean Income', 
                          'Responders': mean_income_responders, 
                          'Non-Responders': mean_income_non_responders}, ignore_index=True)

# Add education distribution
for education_level in education_distribution_responders.index:
    profile = profile.append({'Characteristic': f'Education: {education_level}',
                              'Responders': education_distribution_responders[education_level],
                              'Non-Responders': education_distribution_non_responders.get(education_level, 0)},
                             ignore_index=True)

# Add marital status distribution
for Marital_Status in Marital_Status_distribution_responders.index:
    profile = profile.append({'Characteristic': f'Marital Status: {Marital_Status}',
                              'Responders': Marital_Status_distribution_responders[Marital_Status],
                              'Non-Responders': Marital_Status_distribution_non_responders.get(Marital_Status, 0)},
                             ignore_index=True)

# Add Kidhome distribution
for Kidhome in Kidhome_distribution_responders.index:
    profile = profile.append({'Characteristic': f'Kidhome: {Kidhome}',
                              'Responders': Kidhome_distribution_responders[Kidhome],
                              'Non-Responders': Kidhome_distribution_non_responders.get(Kidhome, 0)},
                             ignore_index=True)
    
# Add Teenhome distribution
for Teenhome in Teenhome_distribution_responders.index:
    profile = profile.append({'Characteristic': f'Teenhome: {Teenhome}',
                              'Responders': Teenhome_distribution_responders[Teenhome],
                              'Non-Responders': Teenhome_distribution_non_responders.get(Teenhome, 0)},
                             ignore_index=True)

# Add NumWebPurchases distribution
for NumWebPurchases in NumWebPurchases_distribution_responders.index:
    profile = profile.append({'Characteristic': f'NumWebPurchases: {NumWebPurchases}',
                              'Responders': NumWebPurchases_distribution_responders[NumWebPurchases],
                              'Non-Responders': NumWebPurchases_distribution_non_responders.get(NumWebPurchases, 0)},
                             ignore_index=True)

# Print the profile
print("Profile of Responders vs. Non-Responders:")
profile
```
